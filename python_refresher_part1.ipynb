## 2. Boolean Algebra

### Explanation
Booleans represent truth values: `True` or `False`. Boolean expressions evaluate to these values.


```python
x = 5
y = 10
print(x < y)  # True
```


Boolean operators:
- `and`: True if both expressions are true
- `or`: True if at least one expression is true
- `not`: Reverses the boolean value

### Example
**Question:** How do you check if x is greater than 5 AND y is less than 20?

```python
x = 7
y = 10
print(x > 5 and y < 20)  # True
```

## 3. If Statements

### Explanation
`if`, `elif`, and `else` statements let you make decisions in your code. They execute blocks of code based on whether a condition is true.

Syntax:

```python
if condition:
    # do something
elif other_condition:
    # do something else
else:
    # do something if none above were true
```


### Example
**Question:** How do you check if a person is an adult?

```python
age = 18
if age < 11: 
    print("You are a child")
elif age <= 18:
    print("You are a teenager")
elif age > 18:
    print("You are an adult")
```
