
## 7. Lists

### Explanation

Lists are ordered collections of items. They can hold any type of data and are mutable.

Useful functions and methods:

- `append(item)`: adds item to the end
- `remove(item)`: removes the first occurrence of item
- `len(list)`: returns number of elements
- `sort()`: sorts the list in place
- `sum(list)`: returns the sum of all numeric elements
- `max(list)`: returns the largest item

### Examples

Looping through a list of fruits and printing them.


```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

Adding and removing items from a list.


```python
foods = ["pizza", "sushi"]
foods.append("tacos")
foods.remove("pizza")
print(foods)  # Output: ['sushi', 'tacos']
```


Sorting a list of numbers.


```python
numbers = [5, 1, 4, 3, 2]
numbers.sort()
print(numbers)  # Output: [1, 2, 3, 4, 5]
```

## 8. Arrays

### Explanation

Arrays are like lists but are more efficient for numerical computations. In Python, you typically use NumPy arrays.

Useful NumPy functions:

- `np.array([...])`: creates an array
- `np.mean(array)`: average of elements
- `np.max(array)`: maximum value
- `np.min(array)`: minimum value
- `np.sum(array)`: sum of all elements

### Examples

Creating an array and finding its mean.


```python
import numpy as np
nums = np.array([10, 20, 30])
print(np.mean(nums))
```


Multiplying all elements of an array.


```python
arr = np.array([1, 2, 3])
print(arr * 2)
```


Creating a 2D array.


```python
matrix = np.array([[1, 2], [3, 4]])
print(matrix)
```

## 9. Dictionaries

### Explanation

Dictionaries store data in key-value pairs. Keys must be unique.

Useful functions and methods:

- `dict.keys()`: returns a list of keys
- `dict.values()`: returns a list of values
- `dict.items()`: returns a list of key-value tuples
- `dict.get(key)`: gets the value for a key safely

### Examples

Looping through a dictionary and printing each subject and its score.


```python
scores = {"math": 90, "english": 85, "science": 95}
for subject, score in scores.items():
    print(subject, score)
```


Accessing values and checking keys.


```python
person = {"name": "Alice", "age": 17}
print(person["name"])
print(person.get("city", "Not specified"))
```


Printing all keys and values.


```python
for key in person.keys():
    print(key)
for value in person.values():
    print(value)
```
