## 1. Functions

### Explanation

Functions let you group code into reusable blocks. You can define your own functions or use built-in ones.

#### Common Built-in Functions:
- `print()`: displays output
- `input()`: reads input from user
- `len()`: gets the length of a list, array, or string
- `sum()`, `max()`, `min()`: perform basic math on collections
- `type()`: tells the type of a variable

### Defining Your Own Functions

Use the `def` keyword:


```python
def greet(name):
    print("Hello", name)
```


### Returning Values

Use `return` to send back a result:


```python
def square(x):
    return x * x
```

## 2. Examples

Function that takes a list of numbers and returns the maximum value.


```python
def find_max(numbers):
    return max(numbers)

print(find_max([5, 8, 2, 10]))
```

Function that receives a number and returns whether it's even or odd.


```python
def is_even(n):
    return n % 2 == 0

print(is_even(7))  # False
```


Function that sums all values in a dictionary.


```python
def sum_scores(scores):
    return sum(scores.values())

grades = {"math": 90, "science": 85}
print(sum_scores(grades))
```

1. Write a function that receives a list of strings and returns a new list with only the strings longer than 4 characters.

```python
def long_than_4(list):
    a = []
    for string in list:
        if len(string) > 4:
            a.append(string)

    return a

```


2. Write a function that receives a NumPy array and returns the average of elements greater than 10.

```python
import numpy as np
def py_recieve(array):
    temp_array = np.array([])
    for number in array:
        if number > 10:
            temp_array = np.append(temp_array,number)

    return  temp_array.mean()
```
